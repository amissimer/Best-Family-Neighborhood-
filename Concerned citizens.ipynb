## School enrollment 

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

schools = pd.read_csv("school-enrollment.csv")
schools.head(10)
#imports the data set of the different schools around pittsburgh and those enrolled and not enrolled 
#split into specfic columns based upon level of school, margin of error, and not enrolled
```

```python
# filter to only show the neighborhoods and the estimated kids enrolled and not enrolled in school 
schools[['Neighborhood', 'Estimate; Enrolled in school:', 'Estimate; Not enrolled in school','Estimate; Total:']]
```

## Diving into the Data

```python
#create a bar graph showing a physical model to depict the enrollment differences between the neighborhoods 
schoolssort = schools.sort_values(by=["Estimate; Enrolled in school:"], ascending=False)

graph = schoolssort.plot.barh(x="Neighborhood",y="Estimate; Enrolled in school:",figsize=(15,schools.shape[0] * 0.3), legend=None)
graph.set_xlabel("Kids Enrolled in School")
graph.set_ylabel("Pittsburgh Neighborhoods")

```

```python
#get an idea of the average number of those enrolled and not enrolled in pittsburgh. This will be used as a comparison
schools[['Neighborhood', 'Estimate; Enrolled in school:', 'Estimate; Not enrolled in school']].describe()
```

```python
# display neighborhoods with above average enrollment 
enrolled = schools[schools['Estimate; Enrolled in school:'] > 988]
enrolled[['Neighborhood', 'Estimate; Enrolled in school:', 'Estimate; Not enrolled in school', 'Estimate; Total:']]
```

```python
enrolled = schools[schools['Estimate; Enrolled in school:'] > 988]
#filter to get rid of neighborhoods that have more than average not enrolled
not_enrolled = enrolled[enrolled['Estimate; Not enrolled in school'] < 2308]
#display results 
not_enrolled[['Neighborhood', 'Estimate; Enrolled in school:', 'Estimate; Not enrolled in school', 'Estimate; Total:']]
```

## Conclusion: 
- According to the data above the "best neighborhood" solely based on enrollment numbers would have to be Central Oakland which in hindsight makes the most sense as it a neighborhood heavily populated by college students. 

## Crime Rates in Pittsburgh Neighborhoods
Investigation of the crime rate in neighborhoods around Pittsburgh to find the best neighborhood for families:

```python
# import libraries
import pandas as pd
import geopandas as gpd
%matplotlib inline

# read csv on crime and confirm it's valid
incidents = pd.read_csv('police_incidents.csv')
incidents.head()
```

## Investigating Data

```python
# organize our dataframe by neighborhood so we can count the number of each 
inc = incidents['INCIDENTNEIGHBORHOOD'].value_counts()

# graph out all data so we can get a general idea
inc.plot(kind = 'bar', figsize=(24,12))
```

```python
# take a look at highest and lowest values
inc.sort_values()
```

```python
# examine only more violent crime
recent = incidents.loc[incidents['HIERARCHY'] < 15]
recent['INCIDENTNEIGHBORHOOD'].value_counts().plot(kind='bar',figsize = (18,12))
```

## Conclusion

- The worst neighborhood for raising a family (i.e. highest crime rate) would be South Side Flats, which is unsuprising considering the high number of bars, nightclubs, and activity that families generally do not partake in.

- The best neighborhood for raising a family when considering the lowest crime rate would be Mt. Oliver, as both boroughs of Mr. Oliver have the lowest number of crime incidents in Pittsburgh since 2016.

## Parks located throughout Pittsburgh neighborhoods
Investigation of the best neighborhood for families in Pittsburgh based off of the number of parks.

```python
#import all of the libraries 
import pandas as pd
import geopandas as gd 
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#imports the data set of parks around pittsburgh
parks_data = pd.read_csv("Parks_Dataset.csv")

#displays a table based off of the data set of the first 5 columns
parks_data.head(5)
```

## Investigating Data 

```python
#filters the data set to only display a table with the name of the park and neighborhood it is in
parks_data[["name", "neighborhood"]]
```

```python
#created a dictionary
dictionary = dict()

#uses a for loop to go through all the rows of the neighborhoods in the data set and if the neighborhood already 
#has a park then it adds one to the existing number and if it does not have that neighborhood in the data set
#then it sets the neighborhood to have one park
for park in parks_data["neighborhood"]:
    if park in dictionary:
        dictionary[park] += 1
    else:
        dictionary[park] = 1

#displays the dictionary
print(dictionary)

```

```python
#changes the dictionary into a series
series = pd.Series(dictionary)
#displays the series
print(series)
```

```python
#sorts the series into highest parks to lowest parks
sorted_series = series.sort_values(ascending=False)
#displays the sorted series
print(sorted_series)
```

```python
#uses the sorted series to make a bar graph that displays each neighborhood with its number of parks from highest to
#lowest 
sorted_series.plot(kind='bar', figsize=(16,12))
```

```python
#displays the top 5 neighborhoods with the highest amount of parks in a bar graph
sorted_series[:5].plot(kind="bar")
```

```python
#displays a box plot using the sorted series to show the average number of parks in each neighborhood is 2
sorted_series.plot(kind="box")
```

# Conclusion #

The best neighborhood for raising a family based off of the numbers of parks would be East Liberty since it has the highest number of parks at 12.
