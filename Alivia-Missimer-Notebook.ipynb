```python
import pandas as pd
import geopandas as gd 
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
parks_data = pd.read_csv("Parks_Dataset.csv")

parks_data.head(25)
```

```python
dictionary = dict()

for park in parks_data["neighborhood"]:
    if park in dictionary:
        dictionary[park] += 1
    else:
        dictionary[park] = 1

print(dictionary)

```

```python
series = pd.Series(dictionary)
print(series)
```

```python
sorted_series = series.sort_values(ascending=False)
    
print(sorted_series)
```

```python
sorted_series.plot(kind='bar', figsize=(16,12))
```
